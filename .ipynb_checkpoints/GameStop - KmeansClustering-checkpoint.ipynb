```python
import pandas as pd
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tokenize import RegexpTokenizer
from string import punctuation
from nltk.corpus import stopwords
import string
import numpy as np
from sklearn import manifold
from sklearn.manifold import MDS
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from nltk.stem import WordNetLemmatizer 
import string
import re
import itertools
#pip install pyspark
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df = pd.read_csv(r"C:\Users\khali\Desktop\McGill\MMA\Winter 2021\INSY670 - Social Media\G2\gamestop_for_network.csv")


df = df.drop_duplicates(subset=['Text'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer


#Term frequency for LDA
tf_mod = CountVectorizer(
    max_df=0.99,
    max_features=500,
    min_df=0.01,
    tokenizer= word_tokenize,
    ngram_range=(1,1))


tf_matrix = tf_mod.fit_transform(df['Text']) 


print ("There are {} reviews and {} terms.".format(
    str(tf_matrix.shape[0]), str(tf_matrix.shape[1])
))
```

```python
#------------------------------------------------------------------------------------------------------
#Elbow method
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(tf_matrix)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#K-means
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA, KernelPCA


kmeans_model = KMeans(n_clusters=2)
kmeans_model.fit(tf_matrix)


kmeans_results = df
clusters = kmeans_model.labels_.tolist()
kmeans_results['cluster'] = clusters


print ("Number of reviews included in each cluster:")
cluster_size = kmeans_results['cluster'].value_counts().to_frame()
cluster_size


pca = KernelPCA(n_components=2)
tfidf_matrix_np=tf_matrix.toarray()
X = pca.fit_transform(tfidf_matrix_np)

xs, ys = X[:, 0], X[:,1]


pca_df = pd.DataFrame(dict(x = xs, y = ys, Cluster = clusters ))
plt.subplots(figsize=(16,9))
sns.scatterplot('x', 'y', data=pca_df, hue='Cluster')
```
